## Plotting Resulting Particle Mass Distributions

Here is the first instance of plotting the standard SSA procedure at the five time snapshots (3,6,9,12,15). We then subsequently begin the process of plotting the the average results of the SSA procedures.

```python
## PURE SSA

runtimes = [3,6,9,12,15]
colors = ['tab:blue', 'tab:orange','tab:green','tab:red','tab:purple']

plotting_fig, ax = plt.subplots(1,1,figsize=(8,6))
fontsize = 20

runs = np.arange(0,1)

all_all_densities = []

for run in runs:
    all_densities = []
    datas = []
    for t in runtimes:
        filename = f'SSA_python_output_nparticles_{110}_{t}_run_{run}.txt'
        data = np.loadtxt(f'/Users/calebpainter/Downloads/Classes/AM207/final_project/outputs/{filename}')
        datas.append(data)

    for n,data in enumerate(datas):
        times = [x[0] for x in data]
        densities = [x[1] for x in data]
        all_densities.append(densities)

        # plt.scatter(times,densities)
        plt.plot(times,densities,label=f'T={n}' if run == 0 else '',color=colors[n],alpha=.4)
        plt.xscale('log')
        plt.yscale('log')

    all_all_densities.append(np.array(all_densities))

for i in range(5):
    times =[x[0] for x in datas[n]]
    time_window = np.array([x[i] for x in all_all_densities])
    avg = averages = np.mean(time_window, axis=0)
    plt.plot(times,avg,color=colors[i])


ax.grid(ls='--',alpha=.5)
ax.set_xlim(1,10**9)
ax.set_ylim(0.02,1)
ax.set_title('SSA',fontsize=fontsize)

ax.tick_params(axis='both', which='major', labelsize=15)

ax.legend(ncol=5,fontsize=13)
ax.set_xlabel("Dimensionless Mass",fontsize=fontsize)
ax.set_ylabel("Mass Density per Bin",fontsize=fontsize)

plt.savefig('/Users/calebpainter/Downloads/Classes/AM207/final_project/poster_plots/SSA.png')
```

```python
def plot_output_distribution(sim_type_num):
# Define runtimes, colors, and other parameters
    runtimes = [3, 6, 9, 12, 15]
    colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']

    plotting_fig, ax = plt.subplots(1, 1, figsize=(15, 4))
    fontsize = 20

    runs = np.arange(0, 30)

    num_particles = [110,100,110]

    # Container for all densities across runs and runtimes
    all_all_densities = []

    for t in runtimes:
        # Collect densities for this runtime across all runs
        runtime_densities = []
        for run in runs:
            npar = num_particles[sim_type_num]
            pfiles = [f'SSA_python_output_nparticles_{npar}_{t}_run_{run}.txt', f'R_mult_leaping_nparticles_{npar}_{100}_reactions_python_output-{t}_run_{run}.txt',f'Tau_leaping_nparticles_{npar}_{0.1}_python_output-{t}_run_{run}.txt']
            pfile = pfiles[sim_type_num]
            data = np.loadtxt(f'/Users/calebpainter/Downloads/Classes/AM207/final_project/outputs/{pfile}')
            times = [x[0] for x in data]
            densities = [x[1] for x in data]
            runtime_densities.append(densities)

        # Compute mean and standard deviation for this runtime
        runtime_densities = np.array(runtime_densities)  # Shape: (runs, len(times))
        mean_densities = np.mean(runtime_densities, axis=0)
        std_densities = np.std(runtime_densities, axis=0)

        # Plot the mean curve with error bars
        if sim_type_num == 0:
            ax.plot(times, mean_densities, label=f'T={t}', color=colors[runtimes.index(t)])
            ax.fill_between(times, mean_densities - std_densities, mean_densities + std_densities, 
                        color=colors[runtimes.index(t)], alpha=0.3)
        else:
            ax.step(times, mean_densities, label=f'T={t}', color=colors[runtimes.index(t)])
            ax.fill_between(times, mean_densities - std_densities, mean_densities + std_densities, step='pre', 
                        color=colors[runtimes.index(t)], alpha=0.3)


    # Customize the plot
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid(ls='--', alpha=0.5)
    ax.set_xlim(1, 10**8)
    ax.set_ylim(0.05, 1)
    if sim_type_num == 1:
        ax.set_title(f'R Leaping, R = {npar}', fontsize=fontsize)
    if sim_type_num == 2:
        ax.set_title(r'$\tau$-Leaping, $\tau$ = 0.1', fontsize=fontsize)

    if sim_type_num == 0:
        ax.set_title(f'SSA', fontsize=fontsize)

    ax.tick_params(axis='both', which='major', labelsize=15)
    ax.legend(ncol=5, fontsize=13)
    ax.set_xlabel("Dimensionless Mass", fontsize=fontsize)
    ax.set_ylabel("Mass Density per Bin", fontsize=fontsize)

    plt.tight_layout()
    # Save the plot
    plt.savefig(f'/Users/calebpainter/Downloads/Classes/AM207/final_project/poster_plots/{sim_type_num}_nice_distribution.png')
    plt.show()
```

```python
### Code from DustPy -- Biernstiel 

def convert(S, m):
    A = np.mean(m[1:]/m[:-1])
    B = 2 * (A-1) / (A+1)
    return S / B

def solution_constant_kernel(t, m, a, S0):
    m0 = m[0]
    N0 = S0 / m0
    return N0 / m0 * 4./(a*N0*t)**2 * np.exp( (1.-m/m0) * 2/(a*N0*t) ) * m**2

def set_constant_kernel(sim, a, S0):
    # Turning off gas evolution by removing integrator instruction
    del(sim.integrator.instructions[1])
    # Turning off gas source to not influence the time stepping
    sim.gas.S.tot[...] = 0.
    sim.gas.S.tot.updater = None
    # Turning off dust advection
    sim.dust.v.rad[...] = 0.
    sim.dust.v.rad.updater = None
    # Turning off fragmentation. Only sticking is considered
    sim.dust.p.frag[...] = 0.
    sim.dust.p.frag.updater = None
    sim.dust.p.stick[...] = 1.
    sim.dust.p.stick.updater = None
    # Setting the constant kernel
    sim.dust.kernel[...] = a
    sim.dust.kernel.updater = None
    # Setting the initial time
    sim.t = 1.e-9
    # Setting the initial dust surface density
    m = sim.grid.m
    A = np.mean(m[1:]/m[:-1])
    B = 2 * (A-1) / (A+1)
    sim.dust.Sigma[...] = sim.dust.SigmaFloor[...]
    sim.dust.Sigma[1, :] = np.maximum(solution_constant_kernel(sim.t, m, a, S0)*B, sim.dust.Sigma[1, :])
    # Updating the simulation object
    sim.update()

def solution_linear_kernel(t, m, a, S0):
    """Analytical solution of the linear collision kernel R(m, m') = a(m + m').
    Initial condition is that only the zeroth mass bin is filled.

    Parameters
    ----------
    t : float
        Time
    m : Field
        Mass grid
    a : float
        Kernel constant
    S0 : float
        Total dust surface density

    Returns
    -------
    Nm2 : Field
        Analytical solution in desired units."""
    m0 = m[0]
    N0 = S0/m0**2
    g = np.exp(-a*S0*t)
    #N = 1./m0**2 * g * np.exp( -m/m0 * ( 1. - np.sqrt(1-g) )**2 ) / ( 2.*np.sqrt(np.pi) * (m/m0)**1.5 * (1.-g)**0.75 )
    N = N0 * g * np.exp( -m/m0 * ( 1. - np.sqrt(1-g) )**2 ) / ( 2.*np.sqrt(np.pi) * (m/m0)**1.5 * (1.-g)**0.75 )
    return N*m**2

def set_linear_kernel(sim, a, S0):
    """Function set the ``DustPy`` simulation object up for the linear collision kernel R(m, m') = a(m + m').

    sim : Frame
        Simulation object
    a : float
        Kernel constant
    S0 : float
        Total dust surface density"""
    # Turning off gas evolution by removing integrator instruction
    del(sim.integrator.instructions[1])
    # Turning off gas source to not influence the time stepping
    sim.gas.S.tot[...] = 0.
    sim.gas.S.tot.updater = None
    # Turning off dust advection
    sim.dust.v.rad[...] = 0.
    sim.dust.v.rad.updater = None
    # Turning off fragmentation. Only sticking is considered
    sim.dust.p.frag[...] = 0.
    sim.dust.p.frag.updater = None
    sim.dust.p.stick[...] = 1.
    sim.dust.p.stick.updater = None
    # Setting the constant kernel
    sim.dust.kernel[...] = a * (sim.grid.m[:, None] + sim.grid.m[None, :])[None, ...]
    sim.dust.kernel.updater = None
    # Setting the initial dust surface density
    sim.dust.Sigma[...] = sim.dust.SigmaFloor[...]
    m = sim.grid.m
    A = np.mean(m[1:]/m[:-1])
    B = 2 * (A-1) / (A+1)
    sim.t = 0.9
    sim.dust.Sigma[1, ...] = np.maximum(solution_linear_kernel(sim.t, m, a, S0) * B, sim.dust.SigmaFloor[1, ...])
    # Updating the simulation object
    sim.update()
```
